```python
questions = [
    "Why are customers complaining about credit cards?",
    "What issues are reported for money transfer services?",
    "What problems do users face with personal loans?",
    "What complaints exist about savings accounts?",
    "Are there fraud-related complaints?"
]
```

```python
results = []

for q in questions:

    answer, sources = answer_question(q)

    results.append({
        "Question": q,
        "Answer": answer,
        "Top Source": sources["document"].iloc[0],
        "Score": sources["score"].iloc[0]
    })

import pandas as pd

pd.DataFrame(results)
```
